```python
from bs4 import BeautifulSoup
import html2text
from tqdm import tqdm
class CustomBeautifulSoupTransformer:
    def transform_document(self, doc):

        if  not doc.page_content:
            return doc
        
        # Parse the document with BeautifulSoup
        soup = BeautifulSoup(doc.page_content, 'html.parser')

        # Extract content within the <article> tag
        article_content = soup.find('article')
        if article_content:
            # Find and process all <a href> tags within the <article> tag
            for a_tag in article_content.find_all('a'):
                a_tag.decompose()
            text = html2text.html2text(f'<p>{article_content}</p>')
            doc.page_content=text
            return doc
        else:
            print(f"Cannot find article in {doc.metadata['source']}")
    def transform_documents(self, docs):
        # Transform each document in the list
        return [self.transform_document(doc) for doc in docs]

# Assuming 'docs' is a list of documents you want to transform
bs_transformer = CustomBeautifulSoupTransformer()
docs_transformed_bs4 = bs_transformer.transform_documents(docs)

# Display the first transformed document
print(docs_transformed_bs4[0][:50])
```
